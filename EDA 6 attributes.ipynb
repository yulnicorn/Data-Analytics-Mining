### This is to show the relationship between the attribute and the age group

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
```

```python
male_df = pd.read_csv('male(imputed).csv')

female_df = pd.read_csv('female(imputed).csv')
```

```python
columns = ['shar_o','sinc_o','fun_o','intel_o','attr_o','amb_o','like_o']
```

```python
processedMaleDF = pd.DataFrame()
for column in columns:
    processedMaleDF[column] = male_df[column].apply(lambda x: int(x))

processedMaleDF['age_o'] = male_df['age_o']
maleAvg = processedMaleDF.groupby('age_o').mean().reset_index()

processedFemaleDF = pd.DataFrame()
for column in columns:
    processedFemaleDF[column] = female_df[column].apply(lambda x: int(x))

processedFemaleDF['age_o'] = female_df['age_o']
femaleAvg = processedFemaleDF.groupby('age_o').mean().reset_index()
```

```python
plt.figure(figsize=(10,7))

plt.plot('age_o','shar_o',color='red',data=maleAvg,label='Shared interest')
plt.plot('age_o','sinc_o',color='orange',data=maleAvg,label='Sincerity')
plt.plot('age_o','fun_o',color='black',data=maleAvg,label='Funny')
plt.plot('age_o','intel_o',color='green',data=maleAvg,label='Intelligence')
plt.plot('age_o','attr_o',color='blue',data=maleAvg,label='Attractiveness')
plt.plot('age_o','amb_o',color='purple',data=maleAvg,label='Ambitious')
plt.plot('age_o','like_o',color='brown',data=maleAvg,label='Like')

plt.xlabel('Age')
plt.ylabel('Score')
plt.title('Male')

plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

plt.savefig('malescore.jpg')
plt.show()

plt.clf()

plt.figure(figsize=(10,7))

plt.plot('age_o','shar_o',color='red',data=femaleAvg,label='Shared interest')
plt.plot('age_o','sinc_o',color='orange',data=femaleAvg,label='Sincerity')
plt.plot('age_o','fun_o',color='black',data=femaleAvg,label='Funny')
plt.plot('age_o','intel_o',color='green',data=femaleAvg,label='Intelligence')
plt.plot('age_o','attr_o',color='blue',data=femaleAvg,label='Attractiveness')
plt.plot('age_o','amb_o',color='purple',data=femaleAvg,label='Ambitious')
plt.plot('age_o','like_o',color='brown',data=femaleAvg,label='Like')

plt.xlabel('Age')
plt.ylabel('Score')
plt.title('Female')

plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

plt.savefig('femalescore.jpg')
plt.show()
```
